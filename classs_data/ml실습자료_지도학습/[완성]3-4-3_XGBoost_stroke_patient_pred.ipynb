```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('../data/healthcare-dataset-stroke-data.csv')
df.head()
```

# 불필요한 Feature 제거

```python
df = df.drop('id', axis=1)
df.head()
```

# Correlation Analysis 수행해보기

```python
corr = df.corr()
plt.figure(figsize=(10, 10));
sns.heatmap(corr,
            vmax=0.8,
            linewidths=0.01,
            square=True,
            annot=True,
            cmap='YlGnBu');
plt.title('Feature Correlation');
```

# 결측치 처리하기

```python
df.isnull().sum()
```

```python
df['bmi'].fillna(df['bmi'].mean(), inplace=True)
```

# String(object) 레이블 encoding하기

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['gender'] = le.fit_transform(df['gender'])
df['ever_married'] = le.fit_transform(df['ever_married'])
df['work_type'] = le.fit_transform(df['work_type'])
df['Residence_type'] = le.fit_transform(df['Residence_type'])
df['smoking_status'] = le.fit_transform(df['smoking_status'])
```

# 트레이닝 데이터 준비하기

```python
y = df['stroke']
X = df.drop('stroke', axis = 1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# Decision Tree로 뇌졸증 예측해보기

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

dt_clf = DecisionTreeClassifier()
dt_clf.fit(X_train, y_train)
y_pred = dt_clf.predict(X_test)

accuracy = accuracy_score(y_pred, y_test)
print('accuracy :', accuracy)
```

```python
from sklearn.metrics import classification_report
# classification_report() 결과 출력
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier()
rf_clf.fit(X_train, y_train)
y_pred = rf_clf.predict(X_test)

accuracy = accuracy_score(y_pred, y_test)
print('accuracy :', accuracy)
```

### [문제] 랜덤 포래스트의 정밀도, 재현율, f1-score 결과 값을 출력하기


```python
from sklearn.metrics import classification_report
# classification_report() 결과 출력
print(classification_report(y_test, y_pred))
```

```python
from xgboost import XGBClassifier

xgb_clf = XGBClassifier(n_estimator=500, learning_rate=0.1, max_depth=5, use_label_encoder=False)
xgb_clf.fit(X_train, y_train)
y_pred = xgb_clf.predict(X_test)

accuracy = accuracy_score(y_pred, y_test)
print('accuracy :', accuracy)
```
